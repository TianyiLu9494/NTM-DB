```python
import os
import pandas as pd
with open('api_species.csv','r') as t:
    t = pd.read_csv(t)
for idx,row in t.iterrows():
    spec_name = row['name']
    print(spec_name)
    if os.path.exists(f"frontend/static/json/grapetree_graph/grapetree_{spec_name}.json"):
        print(f"/static/json/grapetree_{spec_name}.json")
        t.loc[idx,"grapetree_graph_path"] = f"/static/json/grapetree_graph/grapetree_{spec_name}.json"
        
    if os.path.exists(f"frontend/static/json/rs/rs_stacjed/M.{spec_name}_stacked.json"):
        print(f"/static/json/rs/rs_stacjed/M.{spec_name}_stacked.json")
        t.loc[idx,"rs_graph_path"] = f"/static/json/rs/rs_stacjed/M.{spec_name}_stacked.json"
        
    if os.path.exists(f"frontend/static/json/vf/vf_sankey/{spec_name}.json"):
        print(f"/static/json/vf/vf_sankey/{spec_name}.json")
        t.loc[idx,"vf_graph_path"] = f"/static/json/vf/vf_sankey/{spec_name}.json"
    
```
